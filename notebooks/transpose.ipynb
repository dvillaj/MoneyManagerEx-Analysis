```python
import pandas as pd
dataframe = pd.read_excel("https://bit.ly/Tables-xlsx", "table10")
# https://app.box.com/s/3bfwt930wc34bf0wkp6prudsuxr0vk9m

dataframe
```

```python
dataframe \
    .set_index("Symbol") \
    .transpose() \
    .reset_index() \
    .rename( columns = { 'index' : 'Type' }) \
    .rename_axis(None, axis = 'columns') \
    .melt(id_vars="Type",
                 var_name = 'Symbol',
                 value_name = 'Link') \
    .filter(["Symbol", "Type", "Link"])
```
